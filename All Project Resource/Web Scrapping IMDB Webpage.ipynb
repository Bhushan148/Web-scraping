```python
import requests
import requests
from bs4 import BeautifulSoup
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

# Headers to simulate a legitimate browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}

# URL for IMDb Top 250 movies
url = "https://www.imdb.com/chart/top/"

# Make the HTTP request
page = requests.get(url, headers=headers)

# Check if the request was successful
if page.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(page.text, 'html.parser')

    # List to store the extracted data
    movies_list = []

    # Find the container for all movies
    movie_items = soup.find_all('li', class_='ipc-metadata-list-summary-item')

    # Loop over each movie item
    for item in movie_items:
        # Extract movie title
        movie_name_tag = item.find('h3', class_='ipc-title__text')
        movie_name = movie_name_tag.text if movie_name_tag else None

        # Extract movie release year, duration, and other metadata
        metadata_tags = item.find_all('span', class_='cli-title-metadata-item')
        year_tag = metadata_tags[0].text if len(metadata_tags) > 0 else None
        duration_tag = metadata_tags[1].text if len(metadata_tags) > 1 else None

        # Extract movie rating
        rating_tag = item.find('span', class_='ipc-rating-star--rating')
        rating = rating_tag.text if rating_tag else None

        # Extract total ratings
        total_ratings_tag = item.find('span', class_='ipc-rating-star--voteCount')
        total_ratings = total_ratings_tag.text if total_ratings_tag else None

        # Extract image URL
        image_tag = item.find('img', class_='ipc-image')
        image_url = image_tag['src'] if image_tag else None

        # Append the extracted data to the list
        movies_list.append({
            'Movie Name': movie_name,
            'Release Year': year_tag,
            'Duration': duration_tag,
            'Rating': rating,
            'Total Ratings': total_ratings,
        })

    # Convert the list to a DataFrame
    df = pd.DataFrame(movies_list)

    # Display the DataFrame
df
```
