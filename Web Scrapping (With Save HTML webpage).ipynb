```python
# importing libraries
from bs4 import BeautifulSoup
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

# Load the HTML file from the local path
file_path = r'C:\Users\BHUSHAN\Downloads\Mobile Phones _ Mobile Phones Online at Best Prices in India 2024 _ Smartprix.html'
with open(file_path, 'r', encoding='utf-8') as file:
    soup = BeautifulSoup(file, 'html.parser')

# Extract the data
data = []
mobile_elements = soup.find_all('div', class_="sm-product has-tag has-features has-actions")

for mobile in mobile_elements:
    model_name_tag = mobile.find('h2').text.strip()
    price = mobile.find("span", class_="price").text.strip()
    
    rating_span = mobile.find('span', class_='sm-rating')
    rating = rating_span.get('style', '')[9:12] if rating_span else None
    
    hike_tag = mobile.find('span', class_="sm-pdrop").text.strip() if mobile.find('span', class_="sm-pdrop") else None
    
    specs_list = mobile.find('ul', class_='sm-feat specs')
    specs = specs_list.find_all('li')
    
    sim_tag = specs[0].text if len(specs) > 0 else None
    processor_tag = specs[1].text if len(specs) > 1 else None
    ram_tag = specs[2].text if len(specs) > 2 else None
    battery_tag = specs[3].text if len(specs) > 3 else None
    display_tag = specs[4].text if len(specs) > 4 else None
    camera_tag = specs[5].text if len(specs) > 5 else None
    os_tag = specs[6].text if len(specs) > 6 else None
    fm_radio_tag = specs[7].text if len(specs) > 7 else None

    data.append({
        "Model Name": model_name_tag,
        "Price": price,
        "Rating": rating,
        "Price Hike": hike_tag,
        "SIM": sim_tag,
        "Processor": processor_tag,
        "RAM": ram_tag,
        "Battery": battery_tag,
        "Display": display_tag,
        "Camera": camera_tag,
        "OS": os_tag,
        "FM Radio": fm_radio_tag
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Sample data
df
```

```python

# Save the data to a CSV file
output_path = r'C:\Users\BHUSHAN\Downloads\mobile_phones.csv'
df.to_csv(output_path, index=False)
```
